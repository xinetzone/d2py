# {mod}`matplotlib` 关闭事件

示例展示了如何连接在图形关闭时发生的事件。

````python
%matplotlib inline
import matplotlib.pyplot as plt


def on_close(event):
    print('Closed Figure!')

fig = plt.figure()
fig.canvas.mpl_connect('close_event', on_close)

plt.text(0.35, 0.5, 'Close Me!', dict(size=30))
plt.show()
````

```{note}
这个示例展示了 Matplotlib 的交互功能，这在静态文档中将不会显示。请在您的计算机上运行此代码以查看交互性。
```
